Task:4 SALES PREDICTION USING PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/advertising.csv')
df.head()
```

```python
df.tail(5)
```

```python
df.shape
```

```python
df.describe()
x=df.iloc[:,0:-1]
x
y=df.iloc[:,-1]
y
x.iloc[:,0]
df.head(2)
df.corr()
```

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
x_train
x_test
y_train
y_test
```

```python
x_train=x_train.astype('int')
x_test=x_test.astype('int')
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.fit_transform(x_test)
```

Linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

```python
y_pred
y_test
```

Plotting y_test and y_pred data

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,color='red')

```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,color='red',label='Predicted',marker='x')
plt.scatter(y_test,y_pred,color='blue',label='Actual',marker='o')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.show()
```

Check performance of regression

```python
from sklearn.metrics import r2_score, mean_squared_error # Import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
```

Check performance of regression models

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mse
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
mae
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```
